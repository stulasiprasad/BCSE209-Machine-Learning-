```python
import numpy as np

# Step 1: Matrix Formulation (including the bias/intercept column of 1s)
X = np.array([
    [1, 1],
    [1, 2],
    [1, 3],
    [1, 4],
    [1, 5],
    [1, 6]
])

y = np.array([3, 5, 7, 9, 11, 13])

# Step 2 & 3: Transpose and Matrix Multiplication
XtX = X.T @ X
Xty = X.T @ y

# Step 4: Matrix Inversion
XtX_inv = np.linalg.inv(XtX)

# Step 5: Solve for Weights (W = (XᵀX)⁻¹ Xᵀy)
W = XtX_inv @ Xty

print("--- Simple Linear Regression ---")
print(f"Intercept (w0): {W[0]:.1f}")
print(f"Slope (w1):     {W[1]:.1f}")
```

```python
import numpy as np

# Step 1: Matrix Formulation (columns: Intercept, x1, x2)
X = np.array([
    [1, 1, 2],
    [1, 2, 1],
    [1, 3, 4],
    [1, 4, 3],
    [1, 5, 6],
    [1, 6, 5]
])

y = np.array([9, 8, 19, 18, 29, 28])

# Step 2 & 3: Transpose and Matrix Multiplication
XtX = X.T @ X
Xty = X.T @ y

# Step 4: Matrix Inversion
XtX_inv = np.linalg.inv(XtX)

# Step 5: Solve for Weights (W = (XᵀX)⁻¹ Xᵀy)
W = XtX_inv @ Xty

print("\n--- Multiple Linear Regression ---")
print(f"Intercept (w0):  {W[0]:.1f}")
print(f"Feature 1 (w1): {W[1]:.1f}")
print(f"Feature 2 (w2): {W[2]:.1f}")
```

```python
import numpy as np

# Step 1: Matrix Formulation (columns: Intercept, x, x²)
X = np.array([
    [1, 1, 1],
    [1, 2, 4],
    [1, 3, 9],
    [1, 4, 16],
    [1, 5, 25]
])

y = np.array([6, 17, 34, 57, 86])

# Step 2 & 3: Transpose and Matrix Multiplication
XtX = X.T @ X
Xty = X.T @ y

# Step 4: Matrix Inversion
XtX_inv = np.linalg.inv(XtX)

# Step 5: Solve for Weights (W = (XᵀX)⁻¹ Xᵀy)
W = XtX_inv @ Xty

print("\n--- Polynomial Regression ---")
print(f"Intercept (w0): {W[0]:.1f}")
print(f"Linear (w1):    {W[1]:.1f}")
print(f"Quadratic (w2): {W[2]:.1f}")
```

```python
import numpy as np

# Helper function to run gradient descent from scratch
def run_gradient_descent(X, y, alpha=0.1, iterations=100):
    m, n = X.shape
    w = np.zeros(n)  # Initialize weights with zero

    print(f"{"Iteration":<10} | {"Gradients":<30} | {"Weights":<30}")
    print("-" * 78)

    for i in range(1, iterations + 1):
        predictions = X @ w
        errors = predictions - y

        # Calculate gradients: (1/m) * Xᵀ * error
        gradients = (1 / m) * (X.T @ errors)

        # Update weights: w = w - (alpha * gradients)
        w = w - alpha * gradients

        grad_str = ", ".join([f"{g:.4f}" for g in gradients])
        w_str = ", ".join([f"{val:.4f}" for val in w])
        print(f"{i:<10} | [{grad_str}] | [{w_str}]")
    return w
```

```python
# =====================================================================
# 1. SIMPLE LINEAR REGRESSION
# =====================================================================
print("--- 1. SIMPLE LINEAR REGRESSION GRADIENT DESCENT ---")
X_simple = np.array([
    [1, 1],
    [1, 2],
    [1, 3],
    [1, 4],
    [1, 5],
    [1, 6]
])
y_simple = np.array([3, 5, 7, 9, 11, 13])
run_gradient_descent(X_simple, y_simple)
```

```python
# =====================================================================
# 2. MULTIPLE LINEAR REGRESSION
# =====================================================================
print("\n--- 2. MULTIPLE LINEAR REGRESSION GRADIENT DESCENT ---")
X_mlr = np.array([
    [1, 1, 2],
    [1, 2, 1],
    [1, 3, 4],
    [1, 4, 3],
    [1, 5, 6],
    [1, 6, 5]
])
y_mlr = np.array([9, 8, 19, 18, 29, 28])
run_gradient_descent(X_mlr, y_mlr)
```

```python
# =====================================================================
# 3. POLYNOMIAL REGRESSION (DEGREE 2)
# =====================================================================
print("\n--- 3. POLYNOMIAL REGRESSION GRADIENT DESCENT ---")
X_poly = np.array([
    [1, 1, 1],
    [1, 2, 4],
    [1, 3, 9],
    [1, 4, 16],
    [1, 5, 25]
])
y_poly = np.array([6, 17, 34, 57, 86])
run_gradient_descent(X_poly, y_poly)
```
